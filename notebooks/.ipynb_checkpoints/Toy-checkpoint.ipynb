# Spatial-temporal Toy problem
$f_t(\mathbf{u},\lambda) = \lambda_x(\vec{x} - 0.5)^2 - \lambda_y(\vec{y} - 0.5)^2$

$\mathbf{u} = \begin{bmatrix}\vec{x}\\\vec{y}\end{bmatrix}, \qquad \lambda(t) = \begin{bmatrix}\lambda_x(t)\\\lambda_y(t)\end{bmatrix} = \begin{bmatrix}2 + \sin(\omega t)\\ 2 - \sin(\omega t)\end{bmatrix}$

$y_t = f_t(\mathbf{u},\lambda(t)) + \nu_t,\qquad \nu_t\sim\mathcal{N}(0,\mathbf{Q}_t)$

```julia
function f‚Çú(ùêÆ,Œª)
    ùë•,ùë¶ = ùêÆ
    return Œª[1]*(ùë• - 0.5).^2 - Œª[2]*(ùë¶ - 0.5).^2
end
```

```julia
ùëõ‚Çú = 100
Œî·µ§ = 50
ùëõ·µ§ = 25
œâ  = 8œÄ ;
```

```julia
ùë° = linspace(0,1,ùëõ‚Çú)


ùõå = [2 + sin.(œâ.*ùë°),
     2 - sin.(œâ.*ùë°)]

using PyPlot
plot(ùë°, ùõå[1],"-",
     ùë°, ùõå[2],"-")

legend(labels=[L"$\lambda_x$",L"$\lambda_y$"])
xlabel(L"$t$"), ylabel(L"$\lambda$")
title(L"$\omega = 8\pi$")
```

```julia
ùêÆ = (repmat(linspace(0.,1.,Œî·µ§)',Œî·µ§,1), repmat(linspace(0.,1.,Œî·µ§),1,Œî·µ§))

ùêê = 0.1.^2

ùêü = map(Œª -> f‚Çú(ùêÆ,Œª), zip(ùõå[1],ùõå[2]))
ùê≤ = map(Œª -> f‚Çú(ùêÆ,Œª) + chol(ùêê)*randn((Œî·µ§,Œî·µ§)), zip(ùõå[1],ùõå[2]))

; # terminate output
```

```julia
function sample(ùê≤,ùëõ·µ§=10)
    s = rand(1:floor(Int,Œî·µ§),ùëõ·µ§,2)
    s = zip(s[:,1],s[:,2])
    u = zeros((ùëõ·µ§,2))
    u[:,1] = [ùêÆ[1][ùëñ,ùëó] for (ùëñ,ùëó) ‚àà s]
    u[:,2] = [ùêÆ[2][ùëñ,ùëó] for (ùëñ,ùëó) ‚àà s]
    y      = [ùê≤[ùëñ,ùëó]    for (ùëñ,ùëó) ‚àà s]
    return u,y
end    
```

```julia
function kern_rbf(ùë•·µ¢, ùë•‚±º;
                  ls = NaN,
                  v  = NaN,
                  Œ∏::Dict = Dict("lengthscale" => 1.,
                                 "variance"    => 1.))
    
    
    if !isnan(ls)
        Œ∏["lengthscale"] = convert(Float64,ls)
    end
    
    if !isnan(v)
        Œ∏["variance"]    = convert(Float64, v)
    end
    
    if length(ùë•·µ¢) != length(ùë•‚±º)
        println("error: input dimensions incorrect")
        return NaN
    end
    
#     if length(Œ∏["lengthscale"]) < length(ùë•·µ¢)
#         Œ∏new = Dict("lengthscale" => Œ∏["lengthscale"][1]*eye(length(length(ùë•·µ¢))))
#         Œ∏new["variance"] = Œ∏["variance"]
#         Œ∏ = Œ∏new
#     end
    
#     if length(Œ∏["variance"]) < length(ùë•·µ¢)
#         Œ∏["variance"] = 
#         Œ∏new = Dict("variance" => Œ∏["variance"][1]*eye(length(length(ùë•·µ¢))))
#         Œ∏new["lengthscale"] = Œ∏["lengthscale"]
#         Œ∏ = Œ∏new
#     end
    
    ùëò = Œ∏["variance"] .* exp(-norm(ùë•·µ¢ - ùë•‚±º).^2 ./ Œ∏["lengthscale"]^2)
    
    return ùëò
end

function kern_linear(ùë•·µ¢, ùë•‚±º;
                     b = NaN,
                     v = NaN,
                     c = NaN,
                     Œ∏::Dict{String,Float64} = Dict("constant_variance" => 0.,
                                                    "offset"      => 1.,
                                                    "variance"    => 1.))
    
    if !isnan(b)
        Œ∏["constant_variance"] = convert(Float64,b)
    end
    if !isnan(c)
        Œ∏["offset"]            = convert(Float64,c)
    end
    if !isnan(v)
        Œ∏["variance"]          = convert(Float64, v)
    end
    
    ùëò = Œ∏["constant_variance"] 
            + Œ∏["variance"] * (ùë•·µ¢ - Œ∏["offset"]) * (ùë•‚±º - Œ∏["offset"])
    
    return ùëò
end

function kern_periodic(ùë•·µ¢, ùë•‚±º;
                       p  = NaN,
                       ls = NaN,
                       v  = NaN,
                       Œ∏::Dict{String,Float64} = Dict("period"      => 1., 
                                                      "lengthscale" => 0.1,
                                                      "variance"    => 1.))
    if !isnan(p)
        Œ∏["period"] = convert(Float64,p)
    end
    if !isnan(ls)
        Œ∏["lengthscale"] = convert(Float64,ls)
    end
    if !isnan(v)
        Œ∏["variance"]    = convert(Float64, v)
    end
    
    ùëò = Œ∏["variance"] * exp(-(2*sin(œÄ*norm(ùë•·µ¢ - ùë•‚±º)/Œ∏["period"])^2) 
                              / Œ∏["lengthscale"]^2)
    
    return ùëò
    
end

function kern_exponential_ùõæ(ùë•·µ¢, ùë•‚±º;
                            ùõæ  = NaN,
                            ls = NaN,
                            v  = NaN,
                            Œ∏::Dict{String,Float64} = Dict("exponent"    => 1., 
                                                           "lengthscale" => 0.1,
                                                           "variance"    => 1.))
    if !isnan(ùõæ)
        Œ∏["exponent"] = convert(Float64, ùõæ)
    end
    if !isnan(ls)
        Œ∏["lengthscale"] = convert(Float64, ls)
    end
    if !isnan(v)
        Œ∏["variance"]    = convert(Float64, v)
    end
    
    ùëò = Œ∏["variance"] .* exp(-(norm(ùë•·µ¢ - ùë•‚±º)./Œ∏["lengthscale"]).^Œ∏["exponent"])
    
    return ùëò
end
    
function kern_exponential(ùë•·µ¢, ùë•‚±º;
                          ls = NaN,
                          v  = NaN,
                          Œ∏::Dict{String,Float64} = Dict("lengthscale" => 0.1,
                                                           "variance"    => 1.))
    return kern_exponential_ùõæ(ùë•·µ¢, ùë•‚±º, ùõæ = 1., ls = ls, v = v, Œ∏ = Œ∏)
end
```

```julia
function sample_kernel(ùê±; ùëò = kern_rbf, Œ∏ = nothing)
    
    ùëõ‚Çì = length(ùê±)
    
    if Œ∏ == nothing
        ùêä‚Çõ‚Çõ = reshape([ùëò(ùë•·µ¢, ùë•‚±º) for ùë•·µ¢ ‚àà ùê± for ùë•‚±º ‚àà ùê±], ùëõ‚Çì, ùëõ‚Çì)
    else
        ùêä‚Çõ‚Çõ = reshape([ùëò(ùë•·µ¢, ùë•‚±º, Œ∏=Œ∏) for ùë•·µ¢ ‚àà ùê± for ùë•‚±º ‚àà ùê±], ùëõ‚Çì, ùëõ‚Çì)
    end
     
    
    U,S,V = svd(ùêä‚Çõ‚Çõ)
    ùêø = U * diagm(sqrt.(S))
                    
    return ùêø*randn((ùëõ‚Çì,1))
end

```

```julia
x = linspace(-1.5,1.5,400)

params = Dict("lengthscale" => 0.25, "variance" => 1.)
                                
figure()
for _ ‚àà 1:5
    plot(x,sample_kernel(x,ùëò = kern_rbf,Œ∏ = params))
end
title("exponentiated quadratic")    

params = Dict("period" => 1., "lengthscale" => 0.25, "variance" => 1.)
figure()
for i ‚àà [1.,2.,4.]
    params["period"] = i
    plot(x,sample_kernel(x, ùëò = kern_periodic, Œ∏ = params))
end
title("periodic")
legend(labels=[1.,2.,4.])


params = Dict("lengthscale" => 0.5, "variance" => 1.)
figure()
for _ ‚àà 1:5
    plot(x,sample_kernel(x, ùëò = kern_exponential, Œ∏ = params))
end
title("exponential")

params = Dict("exponent" => 1.,"lengthscale" => 0.5, "variance" => 1.)

figure()
for params["exponent"] ‚àà [1.,2.,4.,10.]
    plot(x,sample_kernel(x, ùëò = kern_exponential_ùõæ, Œ∏ = params))
end
title(L"$\gamma$-exponential")
legend(labels=[1.,2.,4.,10.])
```

```julia
function sample_nd_kernel(ùêÆ; ùëò = kern_rbf, Œ∏ = nothing)
    
    ùëõ·µ§ = length(ùêÆ)
    ùëõ‚Çì = length(ùêÆ[1][:])
    
    #iterator (2d only)
    u_itr = zip(ùêÆ[1][:],ùêÆ[2][:])
    
    if Œ∏ == nothing
        ùêä‚Çõ‚Çõ = reshape([ùëò(collect(ùë•·µ¢), collect(ùë•‚±º))
                        for ùë•·µ¢ ‚àà u_itr for ùë•‚±º ‚àà u_itr], ùëõ‚Çì, ùëõ‚Çì)
    else
        ùêä‚Çõ‚Çõ = reshape([ùëò(collect(ùë•·µ¢), collect(ùë•‚±º), Œ∏=Œ∏)
                        for ùë•·µ¢ ‚àà u_itr for ùë•‚±º ‚àà u_itr], ùëõ‚Çì, ùëõ‚Çì)
    end
    
    U,S,V = svd(ùêä‚Çõ‚Çõ)
    ùêø = U * diagm(sqrt.(S))
         
    return ùêø*randn((ùëõ‚Çì,1))
end

```

```julia

function trained_gp(u‚Çõ,y‚Çõ;
                    œÉ¬≤ = 0., # default noise-free
                    ùëò = kern_rbf,
                    Œ∏ = Dict("lengthscale"=>0.25, "variance"=>1.))
    
    ùëõ‚Çõ   = size(u‚Çõ,1)
    
    ùêä‚Çõ‚Çõ  = zeros(ùëõ‚Çõ, ùëõ‚Çõ)
    for ùëñ ‚àà 1:ùëõ‚Çõ
        for ùëó ‚àà 1:ùëõ‚Çõ
            ùêä‚Çõ‚Çõ[ùëñ,ùëó] = ùëò(u‚Çõ[ùëñ,:],u‚Çõ[ùëó,:],Œ∏=Œ∏)
        end
    end
    
#     u·µ¢‚Çú·µ£ = partition(u‚Çõ'[:],size(u‚Çõ,2))
#     ùêä‚Çõ‚Çõ  = reshape([ùëò(collect(ùë•·µ¢), collect(ùë•‚±º))
#                     for ùë•·µ¢ ‚àà u·µ¢‚Çú·µ£ for ùë•‚±º ‚àà u·µ¢‚Çú·µ£], ùëõ‚Çõ, ùëõ‚Çõ)

    return u -> _proto_gp_predict(u, u‚Çõ, y‚Çõ, ùêä‚Çõ‚Çõ + œÉ¬≤*eye(ùëõ‚Çõ), ùëò, Œ∏)
end
                
function _proto_gp_predict(u, u‚Çõ, y‚Çõ, ùêä‚Çõ‚Çõ, ùëò, Œ∏)
    
    ùëõ‚Çõ  = size(u‚Çõ,1)
    ùëõ‚Çä  = size(u,1)
    
    ùêä‚Çõ‚Çä  = zeros(ùëõ‚Çä, ùëõ‚Çõ)
    for ùëñ ‚àà 1:ùëõ‚Çä
        for ùëó ‚àà 1:ùëõ‚Çõ
            ùêä‚Çõ‚Çä[ùëñ,ùëó] = ùëò(u[ùëñ,:],u‚Çõ[ùëó,:],Œ∏=Œ∏)
        end
    end
    
    ùêä‚Çä‚Çä  = zeros(ùëõ‚Çä, ùëõ‚Çä)
    for ùëñ ‚àà 1:ùëõ‚Çä
        for ùëó ‚àà 1:ùëõ‚Çä
            ùêä‚Çä‚Çä[ùëñ,ùëó] = ùëò(u[ùëñ,:],u[ùëó,:],Œ∏=Œ∏)
        end
    end
    
#     u·µ¢‚Çú·µ£  = partition(u‚Çõ'[:],size(u‚Çõ,2))
#     u‚Å∫·µ¢‚Çú·µ£ = partition(u'[:],size(u,2))
#     ùêä‚Çõ‚Çä   = reshape([ùëò(collect(ùë•·µ¢), collect(ùë•‚±º))
#                             for ùë•·µ¢ ‚àà u‚Å∫·µ¢‚Çú·µ£ for ùë•‚±º ‚àà u·µ¢‚Çú·µ£], ùëõ‚Çä, ùëõ‚Çõ)
#     ùêä‚Çä‚Çä   = reshape([ùëò(collect(ùë•·µ¢), collect(ùë•‚±º))
#                             for ùë•·µ¢ ‚àà u‚Å∫·µ¢‚Çú·µ£ for ùë•‚±º ‚àà u‚Å∫·µ¢‚Çú·µ£], ùëõ‚Çä, ùëõ‚Çä)
                                                    
    ùõç = (ùêä‚Çõ‚Çä / ùêä‚Çõ‚Çõ) * y‚Çõ
    ùö∫ = ùêä‚Çä‚Çä - (ùêä‚Çõ‚Çä / ùêä‚Çõ‚Çõ) * ùêä‚Çõ‚Çä'

    return ùõç, ùö∫
end
```

```julia
function plot_prediction(ùêÆ,ùêü,f·µê,f·∂ú=nothing,u=nothing)
    
    Œîx,Œîy = size(ùêÆ[1])
    
    figure()#figsize=(15,8))
    
    if f·∂ú == nothing
        subplot(121)
    else
        subplot(311)
    end
    pcolor(ùêÆ[1],ùêÆ[2],ùêü)
    clim([-1.,1.])
    title("truth")
    colorbar()
    
    if f·∂ú == nothing
        subplot(122)
    else
        subplot(312)
    end
    pcolor(ùêÆ[1],ùêÆ[2],reshape(f·µê,Œîx,Œîy))
    if u != nothing
        plot(u[:,1],u[:,2],".")
    end
    clim([-1.,1.])
    title("gp mean")
    colorbar()

    if f·∂ú != nothing
        subplot(313)
        pcolor(ùêÆ[1],ùêÆ[2],reshape(diag(f·∂ú),Œîx,Œîy))
        if u != nothing
            plot(u[:,1],u[:,2],".")
        end
        title("gp variance")
        colorbar()
    end
end

plot_prediction(ùêÆ,ùêü[1],f·µê,f·∂ú,u)

# figure()
# pcolor(ùêÆ[1],ùêÆ[2],f·∂ú)
# title("variance")
```

```julia
function gp_example()
    x = linspace(0,2œÄ,500)
    f = sin.(x)
    
    œÉ¬≤ = 0.1 ^ 2
    
    i = sort(1 + round.(Int, 500 * rand(10)))
    
    u = x[i]
    y = f[i] + sqrt.(œÉ¬≤) * randn(10)
    
    plot(x,f,"-",u,y,"o")
    
    params = Dict("lengthscale" => 1., "variance" => 0.5)
    
    gp_predict = trained_gp(u, y, ùëò=kern_exponential,
                                  œÉ¬≤=œÉ¬≤,
                                  Œ∏ = params)

    Œº,Œ£ = gp_predict(x)
    
    plot(x,Œº,"--")
   
    fill_between(x,
                 Œº - 1.96*sqrt.(diag(Œ£)), 
                 Œº + 1.96*sqrt.(diag(Œ£)), alpha=0.5)
    
    return Œº, Œ£
end

Œº, Œ£ = gp_example();
```

```julia
function sample_posterior(Œ£)
    
    ùëõ‚Çì = size(Œ£,1)
    
    U,S,V = svd(Œ£)
    ùêø = U * diagm(sqrt.(S))
                    
    return ùêø
end

L = sample_posterior(Œ£);
```

```julia
rmse = zeros(ùëõ‚Çú,1)
vmax = zeros(ùëõ‚Çú,1)
vmed = zeros(ùëõ‚Çú,1)


for t = 1:ùëõ‚Çú
    
    tst_points = [ùêÆ[1][:] ùêÆ[2][:] repmat([convert(Float64,t)],length(ùêÆ[1]),1)]

    f·µê, f·∂ú = gp_predict(tst_points)
    
    if t % 10 == 0 
        plot_prediction(ùêÆ,ùêü[t],f·µê)
    end
    
    rmse[t] = sqrt.(sum((ùêü[t][:] - f·µê).^2))
    vmax[t] = maximum(diag(f·∂ú))
    vmed[t] = median(diag(f·∂ú))
end

#plot(rmse,"-",vmax,"-",vmed,"-")
```
